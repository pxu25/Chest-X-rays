```python
import ipywidgets as wg
from IPython.display import display

# label select
label = wg.SelectMultiple(options=['All','No Finding','Infiltration','Atelectasis','Effusion','Nodule','Pneumothorax',
                                   'Mass','Effusion|Infiltration','Atelectasis|Infiltration','Consolidation'],
                          value  = ['No Finding','Infiltration'],
                     description ='Label:')
label_opt = wg.Select (options     = ['AND', 'OR'],
                            value       = 'OR',
                            description = 'Label Option: ')

# Age slide
age_min   = wg.IntSlider(min=0, max=100, step=1,
                       value = 0,
                       description='Age_min: ')
age_max   = wg.IntSlider(min=0, max=100, step=1,
                       value = 0,
                       description='Age_max: ')
# Gender select
gender = wg.Select (options    = ['Both','F','M'],
                    value      = 'Both',
                    description ='Gender:')
# postion selectat
position = wg.Select(options    = ['Both','PA','AP'],
                     value     = 'Both',
                     description='Position:')

###################################################################

display(label,label_opt,age_min,age_max,gender,position)
```

```python
#import the mysql driver and connect the MariaDB with user name and password
import mysql.connector as mariadb
mariadb_connection = mariadb.connect(host="localhost",
                                     user="user",
                                     passwd="password",
                                     database="X-rays")
cursor = mariadb_connection.cursor()
```

```python
# SQL query
def label_sql(label,label_opt):
    sql = ''
    if label.value[0] != 'All':
        sql = '(FindingLabels = ' + "'" + label.value[0] + "')"
        if len(label.value) > 1:
            sql ="("+(" " +label_opt.value+" ").join(map(lambda x:'FindingLabels = '+ "'"+ x +"'", label.value))+")"
    return sql

def age_sql(age_min, age_max):
    # keep the min less than max
    if age_min.value > age_max.value:
        age_min.value, age_max.value = age_max.value, age_min.value
    sql = ''    
    if age_min.value != 0 or age_max.value != 0:
        sql = '(PatientAge BETWEEN ' + str(age_min.value) + ' AND '+ str(age_max.value)+')'
    return sql

def select_sql(attribute,selection):
    sql = ''
    if selection.value != 'Both':
        sql = attribute + " = " + "'"+selection.value+"'"
    return sql

sql_list = [label_sql(label,label_opt),select_sql('PatientGender',gender),
            select_sql('ViewPosition',position),age_sql(age_min, age_max),]

sql_list_new = [i for i in sql_list if i != '']
    
cntd = ''
if len(sql_list_new) == 1:
    cntd = 'WHRER ' + sql_list_new[0]
if len(sql_list_new) > 1:
    cntd = 'WHERE '+' AND '.join(sql_list_new)

sql      = "SELECT ImageIndex, ImagePath FROM image "
sql      = sql + cntd + ';'

print(sql)
```

```python
# sql query statement

cursor.execute(sql)

##########################################
########Display the results##############
#########################################
print('Your searching paremeters: ',
      '\n    PatienceLabels: ',(' '+label_opt.value+' ').join(label.value), 
      '\n    PatienceAge:    ',(age_min.value,age_max.value), 
      '\n    PatientGender:  ',gender.value, 
      '\n    ViewPosition:   ',position.value)
#import the matplotlib to show the image
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image  as mpimg

n = 0
for i in cursor:
    n +=1
    img    = mpimg.imread(i[1])
    implot = plt.imshow(img, cmap='gray')
    plt.show()
    print('         '+i[0])

print("\n","\n")

print(str(n), " image(s) are found in the database.")

mariadb_connection.close()
    
```
